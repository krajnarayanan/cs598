```python
# generate random weights and bias
def generate_sigma(weights):
    weight_list = []
    bias_list = []
    for i in range(len(weights) - 1):
        w = np.random.randn(weights[i], weights[i+1])
        b = np.random.randn()
        weight_list.append(w)
        bias_list.append(b)
    return weight_list, bias_list
```
